```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
data = pd.read_csv('tested.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Описательная статистика
print(data.describe())

```

```python
# Гистограммы для Age и Fare
plt.figure(figsize=(14, 6))
sns.histplot(data['Age'], bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.histplot(data['Fare'], bins=30, kde=True, color='lightgreen')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

The average fare paid by passengers is approximately $35.58, with significant variability (standard deviation of $55.85). The minimum fare is $0, while the maximum fare reaches $512.33, highlighting the disparity in ticket prices based on class and accommodation.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

The heatmap visualizes the correlation between numerical features in the dataset. A correlation coefficient close to +1 or -1 indicates a strong relationship, while values near 0 indicate weak relationships. For instance, there is a notable positive correlation between Fare and Pclass, suggesting that higher classes tended to pay higher fares.

```python
plt.figure(figsize=(14, 6))
sns.countplot(x='Sex', data=data, palette='pastel')
plt.title('Gender Distribution')
plt.show()
plt.figure(figsize=(14, 6))
sns.countplot(x='Pclass', data=data, palette='pastel')
plt.title('Passenger Class Distribution')
plt.show()
```

This count plot shows the distribution of passengers by gender. It reveals that there were more male passengers compared to female passengers on the Titanic, highlighting a gender disparity in the dataset


```python
mbarked_columns = [col for col in data.columns if 'Embarked_' in col]
print("Embarked columns found:", embarked_columns)

if embarked_columns:
    embarked_counts = data[embarked_columns].sum()
    plt.figure(figsize=(8, 6))
    embarked_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'salmon'])
    plt.title('Embarked Distribution')
    plt.ylabel('')
    plt.show()
else:
    print("No embarked columns found.")
```

The pie chart presents the distribution of embarkation points for passengers. It reveals that the majority of passengers boarded the Titanic in Southampton (Embarked_S), followed by those from Cherbourg (Embarked_C) and Queenstown (Embarked_Q), illustrating the primary departure point.

```python
grouped = data.groupby(['Sex', 'Pclass']).size().unstack()

# Столбчатая диаграмма для группировки
plt.figure(figsize=(10, 6))
grouped.plot(kind='bar', stacked=True, color=['lightblue', 'lightgreen', 'salmon'])
plt.title('Survivor Count by Gender and Passenger Class')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Passenger Class', loc='upper left')
plt.show()
```
